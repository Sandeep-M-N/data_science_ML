## SMOTE(Synthetic Minority Oversampling Technique)

##### SMOTE is a technique used in machine learning to address imbalanced datasets where the minority class has significantly fewer instances than the majority class. SMOTE  involves generating synthetic instances of the minority class by interpolating between existing instances.

```python
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_redundant=0,n_features=2,n_clusters_per_class=1,
                        weights=[0.90],random_state=12)
```

#### Explanation of Parameters
###### Parameter	Description
1. n_samples=1000	Generates 1000 samples (rows) in the dataset.
2. n_features=2	Creates 2 features (columns) for each sample.
3. n_redundant=0	No redundant features (i.e., no extra correlated features).
4. n_clusters_per_class=1	Each class has 1 cluster (group of similar points).
5. weights=[0.90]	90% of the samples belong to one class (imbalanced dataset).
 6. random_state=12	Fixes the random number generator for reproducibility.


```python
import pandas as pd
df1=pd.DataFrame(X,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])
final_df=pd.concat([df1,df2],axis=1)
```

```python
final_df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_df['f1'],final_df['f2'],c=final_df['target'])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
## transform the dataset
oversample=SMOTE()
X,y=oversample.fit_resample(final_df[['f1','f2']],final_df['target'])
```

```python
df1=pd.DataFrame(X,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])
oversample_df=pd.concat([df1,df2],axis=1)
```
